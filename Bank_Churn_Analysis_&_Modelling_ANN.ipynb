# Part 1 - Data Preprocessing

```python
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
# X = dataset.iloc[:, 3:13] --> Selects columns 3 to 12 (inclusive) from the DataFrame and assigns them to X
                              # This is typically where you store your features (input variables) for the model
X = dataset.iloc[:, 3:13]

# Y = dataset.iloc[:, 13] --> Selects column 13 and assigns it to Y. This is usually the target variable (what you're trying to predict – churn in this case)
Y = dataset.iloc[:, 13]
```

```python
print(X)
```

```python
print(Y)
```

```python
 #Create dummy variables

geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

```python
print(geography)
```

```python
print(geography.values)
```

```python
print(gender)
```

```python
## Concatenate the Data Frames

X=pd.concat([X,geography,gender],axis=1)
```

```python
X   # after the concatenated the value of X is being updated
```

```python
## Drop Unnecessary columns

X=X.drop(['Geography','Gender'],axis=1)
```

```python
X   # after the dropping the original 'Geography' & 'Gender' column, the value of X is being updated
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

# Part 2 - Now let's make the ANN!

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, Y_train,validation_split=0.33, batch_size = 10, epochs = 100)
```

```python
# list all data in history

print(model_history.history.keys())
# summarize history for accuracy
plt.plot(model_history.history['accuracy']) # Changed 'acc' to 'accuracy'
plt.plot(model_history.history['val_accuracy']) # Changed 'val_acc' to 'val_accuracy'
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# Part 3 - Making the predictions and evaluating the model

```python
# Predicting the Test set results
Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred > 0.5)

print(Y_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

print(cm)
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(Y_pred,Y_test)

print(score)
```
